## Plot the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states.

```python
master_df
```

```python
pst=master_df.iloc[[6,10,26,30],:].T
```

```python
pst=pst.reset_index()
```

```python
pst.columns=['States','Primary','Secondary','Tertiary','GSDP']
```

```python
pst=pst.drop([0,1])
```

```python
pst['Primary']=pst['Primary'].astype(int)
```

```python
pst['Secondary']=pst['Secondary'].astype(int)
```

```python
pst['Tertiary']=pst['Tertiary'].astype(int)
```

```python
pst['GSDP']=pst['GSDP'].astype(int)
```

```python
pst['%Primary']=round(100*(pst['Primary']/pst['GSDP']),0)
```

```python
pst['%Secondary']=round(100*(pst['Secondary']/pst['GSDP']),0)
```

```python
pst['%Tertiary']=round(100*(pst['Tertiary']/pst['GSDP']),0)
```

## Displaying the percentage contribution of Primary, Secondary and Tertiary with respect to total GSDP of each state

```python
pst
```

```python
pst.plot.bar(x='States',y=['%Primary','%Secondary','%Tertiary'],rot=90,figsize=(15,6))
plt.title('%contribution of Primary, Secondary and Tertiary w.r.t total GSDP')
```

- We can also find the percentage contribution of Primary, Secondary and Tertiary, taken together w.r.t Total GSDP
- Here, PST denotes Primary, Secondary and Tertiary

```python
pst['%PST']=round(100*((pst['Primary']+pst['Secondary']+pst['Tertiary'])/pst['GSDP']),0)
```

```python
pst
```

## Heat Map for percentage contribution of each sector in the total GDP

```python
pst1=pst.loc[:,['States','%Primary','%Secondary','%Tertiary']]
```

```python
pst_pivot=pd.pivot_table(pst1,index='States')
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(pst_pivot,annot=True,cmap="YlGnBu")
plt.title('State-wise Percentage Contribution of each sector to total GDP')
```
